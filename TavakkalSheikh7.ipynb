```python
#Task7: Time Series Analysis
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"\UseC:rs\DELL\Desktop\AAXJ.csv")
print("Dataset Preview")
print(df.head())
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
```

```python
# TASK 1 : TREND ANALYSIS
plt.figure(figsize=(10,5))
plt.plot(
    df["Date"],
    df["Close"]
)
plt.title("Stock Closing Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()
```

```python
# TASK 2 : MOVING AVERAGE FOR TREND
df["Moving_Average"] = (
    df["Close"]
    .rolling(window=30)
    .mean()
)
plt.figure(figsize=(10,5))
plt.plot(
    df["Date"],
    df["Close"],
    label="Actual Price"
)
plt.plot(
    df["Date"],
    df["Moving_Average"],
    label="30-Day Moving Average"
)
plt.legend()
plt.title("Trend Analysis using Moving Average")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
# TASK 3 : SEASONALITY DETECTION
df["Month"] = df["Date"].dt.month
monthly_avg = (
    df.groupby("Month")["Close"]
    .mean()
)
plt.figure(figsize=(8,4))
plt.bar(
    monthly_avg.index,
    monthly_avg.values
)
plt.title("Average Stock Price by Month")
plt.xlabel("Month")
plt.ylabel("Average Closing Price")
plt.show()
```

```python
# OPTIONAL FORECAST
last_price = df["Close"].iloc[-1]
forecast = last_price * 1.02
print("\nCurrent Price:", last_price)
print("Estimated Future Price:", forecast)
```

```python
# INSIGHTS
print("\nTime-Based Insights")
print("Highest Closing Price:")
print(df["Close"].max())
print("\nLowest Closing Price:")
print(df["Close"].min())
print("\nAverage Closing Price:")
print(df["Close"].mean())
```
